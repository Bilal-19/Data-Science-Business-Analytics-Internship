**Importing the Libraries:**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

warnings.filterwarnings('ignore')
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

**Data Collection & Preprocessing:**

```python
# Loading the csv file to a pandas dataframe
student_record = pd.read_csv('student_data.csv')
```

```python
# Print first five rows of the dataframe
student_record.head()
```

```python
# Print last five rows of the dataframe
student_record.tail()
```

```python
# Number of rows and columns in a dataframe
student_record.shape
```

```python
# Checking for missing values in a dataframe
student_record.isnull().sum()
```

```python
# Getting some information about the data
student_record.info()
```

```python
# Getting the statistical measure of the data
student_record.describe()
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

**Separating Descriptive & Target Feature:**

```python
X = student_record.drop(columns = ['Scores'],axis = 1)
Y = student_record['Scores']
```

```python
print(X)
```

```python
print(Y)
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

**Splitting the Training & Testing Data:**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

**Model Training:**

```python
# Loading the linear regression model
model = LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

**Model Evaluation:**

```python
# Prediction on training data
training_data_prediction = model.predict(X_train)
error_score = metrics.r2_score(Y_train, training_data_prediction)
print("R Squared Error:", error_score)
```

```python
# Prediction on test data
test_data_prediction = model.predict(X_test)
error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R Squared Error:", error_score)
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

**Comparing Actual Score vs Predicted Score:**

```python
plt.scatter(Y_test, test_data_prediction,color=['red'])
plt.xlabel('Actual Score')
plt.ylabel('Predicted Score')
plt.title('Actual Score Vs Predicted Score')
plt.show()
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

**Making a Predictive System:**

`What will be the predicted score if the student studies for 9.25 hrs/day?`

```python
study_hours = eval(input("Enter no of study hours per day: "))
array = np.asarray(study_hours)
reshape_array = np.reshape(array, (1,-1))
prediction = model.predict(reshape_array)
print(f"The predicted score if the student studies for {study_hours} hrs/day is {round(prediction[0],3)}%")
```
